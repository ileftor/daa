**Задача 1: Подберите скорость обучения (alpha) и количество итераций.**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[2, 3, 4, 2, 3,
               1, 5, 9, 2, 1],
              [1, 1, 2, 1, 3, 
               0, 5, 10, 1, 2]])
y = [20, 30, 40, 50, 33, 28, 99, 46, 25, 11]

# веса
W = np.array([1, 0.3])

# скорость обучения
alpha = 0.01

# количество итераций
i_num = 3000

n = X.shape[1]
```

```python
# реализуем функцию, определяющую среднеквадратичную ошибку
def mserror(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
for i in range(int(i_num)):
    y_pred = np.dot(W, X)
    err = mserror(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= alpha * (1/n * 2 * np.sum(X[k] * (y_pred - y)))
    if i % 100 == 0:
        print(i, W, err)
```

**Задача 2 (опция): В этом коде мы избавляемся от итераций по весам, но тут есть ошибка, исправьте ее:**

```python
for i in range(int(i_num)):
    y_pred = np.dot(W, X)
    err = mserror(y, y_pred)
    W -= (alpha * (1/n * 2 * np.sum(X * (y_pred - y))))
    if i % 100 == 0:
        print(i, W, err)
```
